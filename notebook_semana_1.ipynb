```python
#ENTRADA: Entra ancho y alto de la madera en metros
ancho = input()
ancho = float(ancho)
alto = input("ingresa el ancho de la ventana en metros")
alto = float(alto)

#PROCESO: Calcula longitud de la madera (ln) en pies
# y el area del vidrio en metros cuadrados
ln = 2 * (ancho + alto) * 3.25
av = 2 * (ancho * alto)

#SALIDA: Sale la longitud de la madera en pies
# el area del vidrio en metros cuadrados
print(" longitud de la madera es", ln, "Pies de longitud")
print("El area del vidrio es", av, "Metros cuadrados")

```
